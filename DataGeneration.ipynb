```python
# Function to generate user-task interaction
import random
import numpy as np
import pandas as pd

def generate_user_task_interaction(users, tasks):
    interactions = []
    skip_percentage = 40 # how many % to skip, 40% of the time, we will skip the interaction
    
    for (_, user) in users.iterrows():
        for (_, task) in tasks.iterrows():

            rand_num = random.randint(0, 100)
            if rand_num < skip_percentage:
                pass

            else:
                if (user["Motivation"] == "Manage Stress" and task["Type"] in ["Steps Tracking", "Start Steps", ]) or \
                    (user["Motivation"] == "Maintain Independence" and task["Type"] in ["Sedentary Benefit", "Sedentary Simple", "Exercise Variety"]) or \
                    (user["Motivation"] == "Feeling Healthy" and task["Type"] in ["MVPA Benefit", "Start MVPA"]): 
                    
                    mean = 0.55
                    std_dev = 0.3
                    completion_prob = np.random.normal(mean, std_dev)

                else:
                    mean = 0.45
                    std_dev = 0.3
                
                    completion_prob = np.random.normal(mean, std_dev)

            # Calculate like probability
            if (user["Motivation"] == "Manage Stress" and task["Type"] in ["Steps Tracking", "Start Steps", ]) or \
                    (user["Motivation"] == "Maintain Independence" and task["Type"] in ["Sedentary Benefit", "Sedentary Simple", "Exercise Variety"]) or \
                    (user["Motivation"] == "Feeling Healthy" and task["Type"] in ["MVPA Benefit", "Start MVPA"]): 

                mean = 0.75
                std_dev = 0.3
                like_prob = np.random.normal(mean, std_dev)
            else:
                mean = 0.45
                std_dev = 0.3
                like_prob = np.random.normal(mean, std_dev)

                completion = completion_prob > 0.5  # True if clipped_num > 0.5, False otherwise
                like = like_prob > 0.5  # True if like_prob > 0.5, False otherwise
           
                interactions.append({"taskid": task["taskid"], "userid": user["userid"], "completion": completion, "like": like})
    return interactions


interactions = generate_user_task_interaction(user, task)

task_user_interaction = pd.DataFrame(interactions)
task_user_interaction = task_user_interaction.reset_index().rename(columns={'index': 'interaction_id'})
task_user_interaction['completion'] = task_user_interaction['completion'].astype(int)

task_user_interaction.to_csv("data/usertasks.csv", index=False)
```
